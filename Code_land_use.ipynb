```python
import pandas as pd
import geopandas as gpd
```

### Load tha grid

```python
grid_gdf = gpd.read_file('Data/grid/grid_geometry.shp')
```

#### Verisk

```python
land_verisk = gpd.read_file('Data/land_cover/land_verisk.gpkg')
land_verisk
```

```python
land_verisk_f = gpd.overlay(grid_gdf, land_verisk, how='intersection')
land_verisk_f['area'] = land_verisk_f['geometry'].area
land_verisk_f['total_area'] = 500 * 500

land_verisk_f['land_use_percentage'] = (land_verisk_f['area'] / land_verisk_f['total_area']) * 100

land_verisk_f = land_verisk_f.groupby(['grid_id', 'landuse_text'])['land_use_percentage'].sum()
land_verisk_f = land_verisk_f.unstack(fill_value=0).reset_index()
land_verisk_f.columns.name = None

land_verisk_f
```

```python
output_file_path = 'Data/land_cover/results/land_verisk.csv'
land_verisk_f.to_csv(output_file_path, index=False, encoding='utf-8')
```

#### Global raster 100m (2019)

```python
import rasterio
from rasterio.features import shapes
import geopandas as gpd
from shapely.geometry import shape

# open geoTIFF document
with rasterio.open('Data/land_cover/global_clip.tif') as src:
    # load the band of data
    band = src.read(1)
    
    # transform the raster data into vector (polygon)
    gdf = (
        {'properties': {'raster_val': v}, 'geometry': s}
        for i, (s, v) in enumerate(shapes(band, transform = src.transform))
    )

    # transform the data into geodataframe
    geoms = list(gdf)
    land_global = gpd.GeoDataFrame.from_features(geoms)
    land_global.set_crs(src.crs, inplace=True)
    
land_global
```

```python
land_global = land_global.to_crs('EPSG:27700')
```

```python
global_label = pd.read_excel('Data/land_cover/global_label.xlsx')
global_label
```

```python
land_global['raster_val'] = land_global['raster_val'].astype(float).astype(int).astype(str)
global_label['Map Code'] = global_label['Map Code'].astype(str)

land_global = land_global.merge(global_label, left_on = 'raster_val', right_on = 'Map Code', how='left')
land_global
```

```python
land_global_f = gpd.overlay(grid_gdf, land_global, how='intersection')
land_global_f['area'] = land_global_f['geometry'].area
land_global_f['total_area'] = 500 * 500

land_global_f['land_cover_percentage'] = (land_global_f['area'] / land_global_f['total_area']) * 100

land_global_f = land_global_f.groupby(['grid_id', 'Land cover class'])['land_cover_percentage'].sum()
land_global_f = land_global_f.unstack(fill_value=0).reset_index()
land_global_f.columns.name = None

land_global_f
```

```python
output_file_path = 'Data/land_cover/results/land_global.csv'
land_global_f.to_csv(output_file_path, index=False, encoding='utf-8')
```
